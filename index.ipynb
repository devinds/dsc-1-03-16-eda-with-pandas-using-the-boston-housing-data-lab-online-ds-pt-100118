# Variable Descriptions

This data frame contains the following columns:

#### crim  
per capita crime rate by town.

#### zn  
proportion of residential land zoned for lots over 25,000 sq.ft.

#### indus  
proportion of non-retail business acres per town.

#### chas  
Charles River dummy variable (= 1 if tract bounds river; 0 otherwise).

#### nox  
nitrogen oxides concentration (parts per 10 million).

#### rm  
average number of rooms per dwelling.

#### age  
proportion of owner-occupied units built prior to 1940.

#### dis  
weighted mean of distances to five Boston employment centres.

#### rad  
index of accessibility to radial highways.

#### tax  
full-value property-tax rate per $10,000.

#### ptratio  
pupil-teacher ratio by town.

#### black  
1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town.

#### lstat  
lower status of the population (percent).

#### medv  
median value of owner-occupied homes in $10000s.

```python
#import pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#set df equal to the file and display information
df=pd.read_csv('train.csv')
df.head()
```

```python
#quickly look at stats for entire dataset
df.info()
df.describe()
```

```python
#Centrality and dispersion for ptratio, the pupil to teacher ratio
print(df['ptratio'].mean())
print(df['ptratio'].median())
print(df['ptratio'].std())

plt.boxplot(df['ptratio'])
plt.title('Pupil to Teacher Ratio, ptratio for Boston Houston Project')
plt.show()

#Boxplot shows that there are some outliers that signify that there are some teachers that have less than 14 pupils
```

```python
#create scatter plot to see if there is correlation between property tax costs and ptratios, particularly smaller classes

plt.scatter(df['ptratio'], df['tax'])
plt.ylabel('y - property-tax rate per $10,000')
plt.xlabel('x - pupil to teacher ratio, ptratio')
plt.title('ptratio by tax rate')
plt.show()

#I made a lot of assumptions here:
#Lower pupil to teacher ratio means smaller class sizes
#Property taxes contribute to school districts and how education systems in the area are funded. 
#Perhaps people who are paying higher property taxes are in areas where the ptratio is lower than average. 
```

```python
#Centrality and dispersion for age
print(df['age'].mean())
print(df['age'].median())
print(df['age'].std())


#The mean and the median age are a 7 year difference, which may mean there's an outlier in the data. 
#Can check for outliers with a boxplot

#boxplot of the age data to check for outliers
plt.boxplot(df['age'])
plt.title('Age Data From Boston Housing Project')
plt.show()

#Data Shows that population skews older and that there is no outliers as suspected previously.
```

```python
#Centrality and dispersion for rm.
print(df['rm'].mean())
print(df['rm'].median())
print(df['rm'].std())

#Histogram of room information
plt.hist(df["rm"], bins=10)
plt.show()

#Histogram shows that most homes in this area have about 6 rooms. Also tells us that smaller homes, (2, 3 bedrooms)
#are not common in this area.
```
